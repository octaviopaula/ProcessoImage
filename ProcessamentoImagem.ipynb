```python
from skimage import io
import numpy as np
import matplotlib .pyplot as plt
```

```python
image = io.imread('https://i.imgur.com/BmlBDBj.png')
```

```python
plt.imshow(image)
```

```python
ImageGray = []
for line in image:
  lineMedia= []
  for r, g, b in line:
    media=int((int(r) + int(g) + int(b)) / 3)
    lineMedia.append(media)
  ImageGray.append(lineMedia)
```

```python
R = 1
M = len(image) - R
N = len(image[0]) - R
```

```python
imageMediana = []
for i in range(R, M):
  newLine = []
  for j in range(R, N):
    filtro = []
    for x in range(i-R, i+R+1):
      for y in range(j-R, j+R+1):
        filtro.append(ImageGray[x][y])
    mediana = int(round(np.median(filtro)))
    newLine.append(mediana)
  imageMediana.append(newLine)
```

```python
plt.imshow(imageMediana, cmap="gray")
```
